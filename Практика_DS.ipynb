```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/differentiated-thyroid-cancer-recurrence/Thyroid_Diff.csv')

# Список категориальных классов
categorical_columns = [ 'Gender', 'Smoking', 'Hx Smoking',
                   'Thyroid Function', 'Physical Examination', 'Adenopathy',
                   'Pathology', 'Focality', 'Risk', 'T', 'N', 'M', 'Stage',
                   'Response', 'Recurred']

# Создание диаграмм счетчиков
plt.figure(figsize=(15, 20))
for i, column in enumerate(categorical_columns):
    plt.subplot(4, 3, i + 1)
    sns.countplot(x=column, data=df)
    plt.title(column)
    plt.xticks(rotation=45)
    plt.xlabel(column)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# создает список ,который содержит названия всех категориальных столбцов в DataFrame
cat_cols=list(df.select_dtypes(include=object).iloc[:,:-1].columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


plt.figure(figsize=(16, len(cat_cols) * 2))

# Создание цветовой карты
cmap = sns.diverging_palette(90, 7, s=75, l=40, n=7, center='light', as_cmap=True)

# Цикл по категориальным столбцам
for i, col in enumerate(cat_cols):
    plt.subplot(len(cat_cols) // 2 + len(cat_cols) % 2, 2, i + 1)
    crosstab_data = pd.crosstab(df[col], df['Recurred'], normalize='index')
    sns.heatmap(data=crosstab_data, annot=True, cmap=cmap, fmt='.2f')
    plt.title(f"{col} recurred")

plt.tight_layout()
plt.show()
```

```python
# подготовка датасета

# Разделение данных для обучения и тестирования
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(df.iloc[:,0:-1], df.iloc[:,-1], test_size= 0.25, random_state=10)
print('X_train shape:',X_train.shape)
print('X_test shape:',X_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape:',y_test.shape)
```

```python
# стандартизирование обучающей и тестовой выборки
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# импорт бибилиотек
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks

# создание экземпляров алгоритмов
sm = SMOTE(sampling_strategy='auto', random_state = 1, k_neighbors=5)
tk = TomekLinks(sampling_strategy='auto')

# создание экземпляра SMOTETomek
# он сначала применяет SMOTE для увеличения меньшинства, а затем применяет Tomek Links для удаления образцов, которые могут быть шумом или находиться на границе между классами
smtk = SMOTETomek(sampling_strategy='auto', smote=sm, tomek=tk, random_state = 1)

# обучающим данным X_train и y_train, чтобы сбалансировать классы
X_over_under, y_over_under = smtk.fit_resample(X_train, y_train)
X_train=X_over_under
y_train=y_over_under
```

```python
# Логистическая регрессия
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
predictions1 = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score

print("Confusion Matrix: \n", confusion_matrix(y_test, predictions1))
print('\n')
print(classification_report(y_test, predictions1))
logreg_acc = accuracy_score(y_test, predictions1)
print("Accuracy of the Logistic Regression Model is: ", logreg_acc)
```

Точность 0.92

```python
# метод опорных векторов
from sklearn.svm import SVC

svc_model = SVC(kernel="rbf")
svc_model.fit(X_train, y_train)
predictions3 = svc_model.predict(X_test)
```

```python
print("Confusion Matrix: \n", confusion_matrix(y_test, predictions3))
print("\n")
print(classification_report(y_test, predictions3))
svm_acc = accuracy_score(y_test, predictions3)
print("Accuracy of SVM model is: ", svm_acc)
```

Точность 0.94

```python
# градиентный бустинг
# создание и обучение модели на тренировочных данных
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
predictions5= gb_model.predict(X_test)
```

```python
# получение отчета о классификации
# вывод общей точности
print("Confusion Matrix: \n", confusion_matrix(y_test, predictions5))
print("\n")
print(classification_report(y_test, predictions5))
gboost_acc= accuracy_score(y_test, predictions5)
print("Accuracy of XGBOOST model is: ", gboost_acc)
```

Точность 0.97

**Вывод:** таким образом, лучшего всего себя показал градиентный бустинг с точностью 0.97 и f1 score 0.94, а хуже всего логистическая регрессия с точностью 0.92 и f1 score 0.82.
